```python
import requests
from bs4 import BeautifulSoup, NavigableString
import datetime 
```

```python
url_base = 'https://www.transfermarkt.com'
season = 2018
df_dict = {
    "name" : [],
    "position" : [],
    "birth_date" : [],
    "nationality" : [],
    "number" : [],
    "season" : [],
    "club" : []
}
```

```python
#Scraping info about a players
for i in range(3):
    season_url = f'{url_base}/serie-a/startseite/wettbewerb/IT1/plus/?saison_id={season}'
    print(f"Scraping season: {season}")
    headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.82 Safari/537.36'}
    response = requests.get(season_url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    table = soup.find('div', class_='large-8 columns').find('table', class_='items')
    links = []
    for team in table.find_all('td', class_='hauptlink'):
        links.append(team.a['href'])
    del links[::2]
    
    print(f"Scraping teams...")
    for link in links:
        print(f"Scraping link: {link}")
        team_url = f'{url_base}{link}'        
        response = requests.get(team_url, headers=headers)
        soup = BeautifulSoup(response.content, 'html.parser')
        
        club = soup.find('div', class_='dataName')
        for player in soup.find_all('tr'): 
            number_class = player.find('div', class_='rn_nummer')
            if number_class != None and number_class.get_text() != '-': #if a player doesn't have a shirt number, it means that he didn't play for the team in that particular season so we can skip him
                df_dict['number'].append(number_class.get_text())
                name_class = player.find('td', class_='posrela')
                if name_class != None:
                    df_dict['name'].append(name_class.find('table', class_='inline-table').find('span').get_text())
                    df_dict['position'].append(name_class.find('table', class_='inline-table').tr.find_next_sibling().get_text())
                    df_dict['birth_date'].append(name_class.find_next_sibling().get_text()[:-5])
                    df_dict['nationality'].append(name_class.find_next_sibling().find_next_sibling().img.get('alt'))
                    df_dict['season'].append(season)
                    df_dict['club'].append(club.get_text().strip())
    season +=1
    
df_dict['number'] = [int(i) for i in df_dict['number']] #convert string to int 
df_dict['birth_date'] = [datetime.datetime.strptime(d, '%b %d, %Y').date() for d in df_dict['birth_date']] #convert string to date
```

```python
import re

#Obtain list of ids of teams for each of 3 seasons 
clubs_dict = {}
season = 2018
for i in range(3):
    clubs_dict[str(season+i)] = []
    season_url = f'https://www.espn.com/soccer/standings/_/league/ITA.1/season/{season+i}'
    response = requests.get(season_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    for club in soup.find('tbody', class_='Table__TBODY').find_all('tr'):
        clubs_dict[str(season+i)].append(re.search('[\d]+', club.td.div.a['href']).group())
clubs_dict  
```

```python
#Iterate through all of the teams match history for the conrete years to get the match ids and match dates
# match_ids = []
match_ids = {}
for season, club_ids in clubs_dict.items():
    print(f'Season: {season}')
    for club_id in club_ids: 
        print(f'Club: {club_id}')
        club_season_results_url = f'https://www.espn.com/soccer/team/results/_/id/{club_id}/season/{season}'
        response = requests.get(club_season_results_url)
        soup = BeautifulSoup(response.content, 'html.parser')
        
        for match in soup.find('section', class_='Card').find_all('tr', class_='Table__TR Table__TR--sm Table__even'): #Get all matches played in the season for a team
            if re.search('[\d]+', match.a['href']).group() == club_id and match.select("td:nth-of-type(6)")[0].get_text() == 'Italian Serie A': #Get matches that were played only at home and filter out only 'Italian Serie A' matches..
                year = match.find_parent('div', class_='ResponsiveTable Table__results').div.get_text()[-4:]
                match_link = match.select("td:nth-of-type(3)")[0].a.find_next_sibling()['href'] #Get match id from the link and append it to the list of match ids 
                match_id = re.search('[\d]+', match_link).group()
#                 match_ids.append(re.search('[\d]+', match_link).group())
                date = match.td.get_text()[5:] + ' ' + year
                match_ids[str(match_id)] = datetime.datetime.strptime(date, '%b %d %Y').date()
#                 print(re.search('[\d]+', match_link).group())
match_ids
```

```python
import time
def get_soup(id):
    match_url = f'https://www.espn.com/soccer/commentary?gameId={id}'
    response = requests.get(match_url)
    soup = BeautifulSoup(response.content, 'html.parser')
#     time.sleep(1)
    
    if soup.find('h1', class_='Error404__Title') != None:
        return soup, False
    return soup, True
```

```python
def get_player_fk(player_name, player_number, season, club_fk):
    player_fk = df[(df['name'] == player_name) & (df['number'] == player_number) & (df['season'] == season)]
    if player_fk.empty:
#         print("Name inconsistency.. searching using only number of the player")
        player_fk = df[(df['club'] == club_fk) & (df['number'] == player_number) & (df['season'] == season)]
        if player_fk.empty:      
#             print("Name inconsistency.. searching using part of the name")
            name = player_name.split()
            name.reverse() #TODO
            player_fk = df[(df['name'].str.contains('|'.join(name))) & (df['number'] == player_number) & (df['season'] == season)]
            if player_fk.empty:
                player_fk = None
            else:
                player_fk = player_fk.index[0]
        else:
            player_fk = player_fk.index[0]
    else:
        player_fk = player_fk.index[0]
    return player_fk
```

```python
soup = None
date = None
checkpoint = 0
for match_id, date in match_ids.items():
    print(checkpoint)
#     if(checkpoint < 199):
#         checkpoint+=1
#         continue
#     if(checkpoint == 3):
#         break
    checkpoint+=1
    print(match_id, date)
    soup, flag = get_soup(match_id)
    
    while not flag:
        print("Server error, retrying..")
        soup, flag = get_soup(match_id)
    
    table = soup.find('div', class_='accordion active')
    teams = soup.find_all('div', class_='team-info-wrapper')
    teams = [teams[0].span.get_text(), teams[1].span.get_text()]
    match_fk = matches_data[(matches_data['HomeTeam'] == teams[0]) & (matches_data['AwayTeam'] == teams[1]) \
                   & ((matches_data['Date'] == np.datetime64(date)) | \
                      ((matches_data['Date'].dt.day == np.datetime64(date).astype(object).month) & (matches_data['Date'].dt.month == np.datetime64(date).astype(object).day))) ].index[0]
    season = int(soup.find('div', class_='game-details header').get_text().strip()[:4])
    home_club_fk = clubs_df[clubs_df['club_name'] == teams[0]].index[0]
    away_club_fk = clubs_df[clubs_df['club_name'] == teams[1]].index[0]
    
    def get_number_string(event):
        minute = event.td.get_text().replace("'","")
        string = event.td.find_next_sibling().find_next_sibling().get_text().strip()
        return minute, string

    def add_row_to_dict(event, minute, player):
        match_details['minute'].append(minute)
        match_details['event'].append(event)
        match_details['match_id'].append(match_fk)
        player_fk = df[(df['name'] == player) & (df['season'] == season) ]
        if player_fk.empty:
            name = player.split()
            name.reverse()
            player_fk = df[(df['name'].str.contains('|'.join(name))) & (df['season'] == season) ]#& ((df['club'] == home_club_fk) | (df['club'] == away_club_fk))]
            if player_fk.empty:
                player_fk = None
            else:
                player_fk = player_fk.index[0]
        else:
            player_fk = player_fk.index[0]
#         print(player_fk)
#         team_player_fk = match_players_df[(match_players_df['match_id'] == match_fk) & ((match_players_df['club_id'] == home_club_fk) | (match_players_df['club_id'] == home_club_fk)) & (match_players_df['player_id'] == player_fk)]
#         print(team_player_fk)
        team_player_fk = match_players_df[(match_players_df['match_id'] == match_fk) & (match_players_df['player_id'] == player_fk)]
        if team_player_fk.empty:
            team_player_fk = None
        else:
            team_player_fk = team_player_fk.index[0]
        match_details['team_player_id'].append(team_player_fk)


    for event in table.find_all('tr', attrs={"data-type": "yellow-card"}):
        minute, text_string = get_number_string(event)
        player = re.search('^(.*) \(', text_string).group(1)
        add_row_to_dict("yellow card", minute, player)

    for event in table.find_all('tr', attrs={"data-type": "red-card"}):
        minute, text_string = get_number_string(event)
        player = re.search('^(.*) \(', text_string).group(1) 
        add_row_to_dict("red card", minute, player)

    for event in table.find_all('tr', attrs={"data-type": "corner-kick"}):
        minute, text_string = get_number_string(event)
        player = re.search('^.*by (.*).$', text_string)
        if player == None:
            continue
        else:
            player = player.group(1)
        add_row_to_dict("corner", minute, player)

    for event in table.find_all('tr', attrs={"data-type": "goal"}):
        minute, text_string = get_number_string(event)
        goal_player = re.search('^.*\. (.*) \(', text_string).group(1)
        add_row_to_dict("goal", minute, goal_player)
    #     assist_player = re.search('Assisted by ([A-Z]([a-z\']+|\.)(?:\s+[A-Z]([a-z\']+|\.))*(?:\s+[a-z\'][a-z\'\-]+){0,2}\s+[A-Z]([a-z\']+|\.))', text_string)
        assist_player = re.search('Assisted by (.*?)((?= following)|(?= with)|(?=\.))', text_string)
        if assist_player is not None:
            assist_player = assist_player.group(1)
            add_row_to_dict("assist", minute, assist_player)

    for event in table.find_all('tr', attrs={"data-type": "own-goal"}):
        minute, text_string = get_number_string(event)
        player = re.search('by (.*?), ', text_string).group(1)
        add_row_to_dict("own goal", minute, player)
        
    for event in table.find_all('tr', attrs={"data-type": "offside"}):
        minute, text_string = get_number_string(event)
        player = re.search('but (.*?) is', text_string)
        if player == None:
            continue
        else:
            player = player.group(1)
        add_row_to_dict("offside", minute, player)

    for event in table.find_all('tr', attrs={"data-type": "substitution"}):
        minute, text_string = get_number_string(event)
        player = re.search('\. (.*) replaces (.*)(?=\.)', text_string)
        sub_on = player.group(1)
        add_row_to_dict("sub on", minute, sub_on)
        if(player.group(2).endswith('injury')):
            sub_off = player.group(2).rsplit(' ', 4)[0]
        else:
            sub_off = player.group(2)
        add_row_to_dict("sub off", minute, sub_off)
    #     print(event_type, minute, sub_on, sub_off)

```
